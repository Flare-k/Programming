# 서울시 상권 분석하기

### 1. 주제 선정 이유 

증조할아버지 이래로 외할머니, 어머니, 그리고 본인까지 4대째 성수동에서 태어나고 생활했다. 그러나 성수동의 재개발이 확정이 된 이후로 2017년도 약 1세기만에 '동'을 떠나게 되었다. 비교적 가까운 거리에 있는 자양동으로 이사를 가게 되었으면서도 계속 다니던 성수동에서의 교회도 옮기지 않을 뿐더러 최근 3월까지 성수동에 있는 카페에서 약 2년 동안 아르바이트를 했었다. 그만큼 아직까지도 성수동에 대한 애착이 크다.<br>
그런 성수동이 최근들어 여러 매체에서 언급될 뿐만 아니라 수제화, 성수동 카페거리 등으로 이전과는 달리 급속도로 성장하게 되어 사람들 사이에서 그야말로 '핫플레이스'가 되었다. 이전부터 이미 유명했던 홍대, 혜화, 건대 등과 견주어봤을 때 얼마나 사람들 사이에서 나의 옛 동네가 인기몰이를 하는지 궁금해졌다. 그렇게 성수동에서의 작은 관심으로 시작하여 <br>
1) 실제로 사람들의 관심을 많이 받는 동네는 어디인지,<br>
2) 평일과 주말은 확연하게 사람들의 분포도가 달라질텐데, 특히 주말이기에도, 또 평일이라 하기에도 모호한 '불금'에는 선호 지역이 달라진다면 어떻게 달라지는지,<br>
또한 예전부터 개인적인 궁금증이 있었는데, 본인이 항상 놀러 나갈 때에 하는 말이 있다. 홍대 쪽은 사람들이 너무 많아 이리 치이고 저리 치이게 되는 반면, 신촌은 어느 정도 사람 수가 정제되어 있어 정돈된 느낌과 평화로움을 느낄 수 있기에 신촌 쪽을 보다 선호한다고 이야기하곤 한다. 하지만 이에 대해 신촌이나 어디나 별반 다를 게 없다고 하는 사람들도 종종 있었다. 과연 신촌은 사람들이 일반적으로 알고 있는 번화가에 비해서 사람 수가 적은 지, 아니면 단순 본인의 느낌이 그런 것인지 이 또한 추가적으로 사실로서 다가가고자 한다.

### 2. 가설 정의


실제로 사람들의 관심을 많이 받는 동네는 어디이며,<br>
평일과 주말의 분포도 및 '불금'은 주로 어느 곳을 선호하는지,<br>
주말에 시간대 별로 특정 지역에 몰리는 경우가 있는지(그렇다면 사람 많은 걸 싫어하는 본인에게 주말 약속 장소로 적합한 곳은 어디일지)<br>
남녀 성비가 특별히 갈리는 지역이 있다면 어느 지역인지(창업 관련에 있어 도움이 될 것이다)

## 다루게 될 모든 데이터에 관하여

url에서 startnum과 endnum이 각각 1, 5일 때에는 위와 같이 다섯 개의 거리에 대해서만 출력되었다.<br>

```python

startnum=1
endnum=1000
Commerceinfor={}

codelist=[]
codenamelist=[]
totalnumberlist=[]
maletotallist=[]
femaletotallist=[]
timefrom9list=[]
man20monlist=[]
man20tuelist=[]
man20wedlist=[]
man20thulist=[]
man20frilist=[]
man20satlist=[]
man20sunlist=[]
woman20monlist=[]
woman20tuelist=[]
woman20wedlist=[]
woman20thulist=[]
woman20frilist=[]
woman20satlist=[]
woman20sunlist=[]
while endnum<=2000:
    url='http://openapi.seoul.go.kr:8088/5a61756e6e6a756136336776525471/xml/InfoTrdarFlpop/'+str(startnum)+'/'+str(endnum)+'/201810'
    #str(startnum)+'/'+str(endnum)
    req=requests.get(url)
    html=req.text
    soup=BeautifulSoup(html, 'html.parser')
    
    codenumber=soup.find_all('trdar_cd')#trdar_cd = 000001, 000002, ...
    codename=soup.find_all('trdar_cd_nm')
    totalnumber=soup.find_all('tot_flpop_co')
    maletotal=soup.find_all('ml_flpop_co')
    femaletotal=soup.find_all('fml_flpop_co')
    timefrom9=soup.find_all('tmzon_6_flpop_co')
    man20mon=soup.find_all('mag_20_montm_6_flpop_co')
    man20tue=soup.find_all('mag_20_tuetm_6_flpop_co')
    man20wed=soup.find_all('mag_20_wedtm_6_flpop_co')
    man20thu=soup.find_all('mag_20_thutm_6_flpop_co')
    man20fri=soup.find_all('mag_20_fritm_6_flpop_co')
    man20sat=soup.find_all('mag_20_sattm_6_flpop_co')
    man20sun=soup.find_all('mag_20_suntm_6_flpop_co')
    woman20mon=soup.find_all('fag_20_montm_6_flpop_co')
    woman20tue=soup.find_all('fag_20_tuetm_6_flpop_co')
    woman20wed=soup.find_all('fag_20_wedtm_6_flpop_co')
    woman20thu=soup.find_all('fag_20_thutm_6_flpop_co')
    woman20fri=soup.find_all('fag_20_fritm_6_flpop_co')
    woman20sat=soup.find_all('fag_20_sattm_6_flpop_co')
    woman20sun=soup.find_all('fag_20_suntm_6_flpop_co')
    
  
    for code in codenumber:
        codelist.append(code.text)#trdar_cd 따위의 값이 안나옴
    for code in codename:
        codenamelist.append(code.text)
    for code in totalnumber:
        totalnumberlist.append(code.text)
    for code in maletotal:
        maletotallist.append(code.text)
    for code in femaletotal:
        femaletotallist.append(code.text)
    for code in man20mon:
        man20monlist.append(code.text)
    for code in man20tue:
        man20tuelist.append(code.text)
    for code in man20wed:
        man20wedlist.append(code.text)
    for code in man20thu:
        man20thulist.append(code.text)
    for code in man20fri:
        man20frilist.append(code.text)
    for code in man20sat:
        man20satlist.append(code.text)
    for code in man20sun:
        man20sunlist.append(code.text)
    for code in woman20mon:
        woman20monlist.append(code.text)
    for code in woman20tue:
        woman20tuelist.append(code.text)
    for code in woman20wed:
        woman20wedlist.append(code.text)
    for code in woman20thu:
        woman20thulist.append(code.text)
    for code in woman20fri:
        woman20frilist.append(code.text)
    for code in woman20sat:
        woman20satlist.append(code.text)
    for code in woman20sun:
        woman20sunlist.append(code.text)
    
    startnum+=1000
    endnum+=1000 
```

```python
Commerceinfor['Code'] = codelist
Commerceinfor['Name'] = codenamelist
Commerceinfor['Total Number'] = totalnumberlist
Commerceinfor['Male Total Number'] = maletotallist
Commerceinfor['Female Total Number'] = femaletotallist
Commerceinfor['남성 20대 - 월']=man20monlist
Commerceinfor['남성 20대 - 화']=man20tuelist
Commerceinfor['남성 20대 - 수']=man20wedlist
Commerceinfor['남성 20대 - 목']=man20thulist              
Commerceinfor['남성 20대 - 금']=man20frilist
Commerceinfor['남성 20대 - 토']=man20satlist
Commerceinfor['남성 20대 - 일']=man20sunlist              
Commerceinfor['여성 20대 - 월']=woman20monlist              
Commerceinfor['여성 20대 - 화']=woman20tuelist
Commerceinfor['여성 20대 - 수']=woman20wedlist
Commerceinfor['여성 20대 - 목']=woman20thulist
Commerceinfor['여성 20대 - 금']=woman20frilist
Commerceinfor['여성 20대 - 토']=woman20satlist
Commerceinfor['여성 20대 - 일']=woman20sunlist
              

df = pd.DataFrame(Commerceinfor)
df = df[['Code','Name', 'Total Number','Male Total Number','Female Total Number',
'남성 20대 - 월','남성 20대 - 화','남성 20대 - 수','남성 20대 - 목','남성 20대 - 금','남성 20대 - 토','남성 20대 - 일',
'여성 20대 - 월','여성 20대 - 화','여성 20대 - 수','여성 20대 - 목','여성 20대 - 금','여성 20대 - 토','여성 20대 - 일']]
df
```

### 남녀 요일별 합계, 즉 요일별 21시 이후의 유동인구 구하기

```python
all_mon=[]
all_tue=[]
all_wed=[]
all_thu=[]
all_fri=[]
all_sat=[]
all_sun=[]
weekday=[]
weekend=[]
total20=[]
for i in range(len(man20monlist)):
    man20monlist[i]=int(man20monlist[i])
    man20tuelist[i]=int(man20tuelist[i])
    man20wedlist[i]=int(man20wedlist[i])
    man20thulist[i]=int(man20thulist[i])
    man20frilist[i]=int(man20frilist[i])
    man20satlist[i]=int(man20satlist[i])
    man20sunlist[i]=int(man20sunlist[i])
    woman20monlist[i]=int(woman20monlist[i])
    woman20tuelist[i]=int(woman20tuelist[i])
    woman20wedlist[i]=int(woman20wedlist[i])
    woman20thulist[i]=int(woman20thulist[i])
    woman20frilist[i]=int(woman20frilist[i])
    woman20satlist[i]=int(woman20satlist[i])
    woman20sunlist[i]=int(woman20sunlist[i])
    
for i in range(len(man20monlist)):
    all_mon.append(man20monlist[i]+woman20monlist[i])
    all_tue.append(man20tuelist[i]+woman20tuelist[i])
    all_wed.append(man20wedlist[i]+woman20wedlist[i])
    all_thu.append(man20thulist[i]+woman20thulist[i])
    all_fri.append(man20frilist[i]+woman20frilist[i])
    all_sat.append(man20satlist[i]+woman20satlist[i])
    all_sun.append(man20sunlist[i]+woman20sunlist[i])
    weekday.append((all_mon[i]+all_tue[i]+all_wed[i]+all_thu[i])/4)
    weekend.append((all_sat[i]+all_sun[i])/2)
    total20.append(weekday[i]+weekend[i]+all_fri[i])
    
```

```python
total20rank=sorted(total20, reverse=True)
total20rankindex=[]
for i in range(len(total20rank)):
    for j in range(len(total20)):
        if len(total20rankindex)==100:
            break
        if total20rank[i]==total20[j]:
            total20rankindex.append(j)

print(total20rankindex)
```
