# 모듈을 이용한 은행 프로그램

```python
from bankteller import BankTeller
 

menu_list = '''\t*** XXX 온라인 은행 ***
 
1. 고객 정보 등록
2. 고객 정보 조회
3. 신규 계좌 개설
4. 계좌 정보 조회
5. 입출금 업무
6. 프로그램 종료
'''
 
def view_register_customer(cls):
    while True:
        print ('\t ### 1. 고객 정보 등록 메뉴')
        _id = input('- 1. 고객 아이디 입력 >>> ')
         
        if _id == '':
            print('고객 아이디는 필수 항목 입니다. 다시 입력해 주세요')
            continue
        elif cls.existCustomerId(_id):
            print('입력하신  ID(%s) 가 이미 있습니다.' % _id)
            continue
         
        pwd = input('- 2. 고객 패스워드 입력 >>> ')
         
        if pwd == '':
            print('고객 패스워드는 필수 항목 입니다. 다시 입력해 주세요')
            continue
                 
        name = input('- 3. 고객 이름 입력 >>> ')
        
        if name == '':
            print('고객 이름은 필수 항목 입니다. 다시 입력해 주세요')
            continue
         
        try:
            age = int(input('- 4. 고객 나이 입력 >>> '))
        except:
            age = 0
         
        gender = input('- 5.고객 성별 입력 (M=남자, F=여자) >>> ')        
 
        if gender == '' or (gender != 'M' and gender != 'F'):
            gender = 'N'
        if cls.registerCustomer(_id, pwd, name, age, gender):
            break

def view_search_customer(cls):
    while True:
        print ('\t ### 2. 고객 정보 조회')
        _id = input('- 1. 고객 아이디 입력 >>> ')
         
        if _id == '':
            print('고객 아이디는 필수 항목 입니다. 다시 입력해 주세요')
            continue
         
        pwd = input('- 2. 고객 패스워드 입력 >>> ')
         
        if pwd == '':
            print('고객 패스워드는 필수 항목 입니다. 다시 입력해 주세요')
            continue
         
        cls.searchCustomer(_id, pwd)
        break

def view_create_account(cls):
    while True:
        print ('\t ### 3. 신규 계좌 개설')
        _id = input('- 1. 고객 아이디 입력 >>> ')
         
        if id == '':
            print('고객 아이디는 필수 항목 입니다. 다시 입력해 주세요')
            continue  
 
        pwd = input('- 2. 고객 패스워드 입력 >>> ')
         
        if pwd == '':
            print('고객 패스워드는 필수 항목 입니다. 다시 입력해 주세요')
            continue
         
        try:
            balance = int(input('- 3. 입금 금액 >>> '))
        except:
            balance = 0
            print('입력된 금액이 잘못되었거나 업습니다.')
         
         
        cls.createAccount(_id, pwd, balance)
        break
    
def view_search_account(cls):
    while True:
        print ('\t ### 4. 계좌 정보 조회')
        _id = input('- 1. 고객 아이디 입력 >>> ')
        if _id == '':
            print('고객 아이디는 필수 항목 입니다. 다시 입력해 주세요')
            continue
         
        pwd = input('- 2. 고객 패스워드 입력 >>> ')
         
        if pwd == '':
            print('고객 패스워드는 필수 항목 입니다. 다시 입력해 주세요')
            continue
         
        cls.searchAccount(_id, pwd)
        break;

def view_process_balance(cls):
    while True:
        print ('\t ### 5. 입출금 업무')
        _id = input('- 1. 고객 아이디 입력 >>> ')
         
        if _id == '':
            print('고객 아이디는 필수 항목 입니다. 다시 입력해 주세요')
            continue
         
        pwd = input('- 2. 고객 패스워드 입력 >>> ')
         
        if pwd == '':
            print('고객 패스워드는 필수 항목 입니다. 다시 입력해 주세요')
            continue
         
        while True:
            try:
                choice = int(input('메뉴 입력(1 : 입금, 2 : 출금, 3 : 종료) >>> '))
                input_value = 0
                 
                if choice == 1: # 입금
                    input_value = int(input('계좌에 입금할 금액을 입력하세요 >>>  '))
                    cls.depositAccount(_id , pwd, input_value)
                    break
                elif choice == 2: # 출금
                    input_value = int(input('계좌에 출금할 금액을 입력하세요 >>>  '))
                    cls.withdrawAccount(_id , pwd, input_value)
                    break
                elif choice == 3:
                    break
                else:
                    print('*** 올바른 메뉴를 다시 입력해주세요')
                    continue
                                
            except:
                print('*** 올바른 메뉴를 다시 입력해주세요')
         
        break 

def view_main():
    bank_teller = BankTeller('LT1234', 'XXX Teller', 1, 'C')
 
    while True:
        print(menu_list)
        try:
            choice = int(input('메뉴 입력(1~6) : '))
            if choice == 6 : # 6. 프로그램 종료
                break;            
            elif choice == 1: # 1. 고객 정보 등록
                view_register_customer(bank_teller)
            elif choice == 2: # 2. 고객 정보 조회
                view_search_customer(bank_teller)
            elif choice == 3: # 3. 신규 계좌 개설
                view_create_account(bank_teller)
            elif choice == 4: # 4. 계좌 정보 조회
                view_search_account(bank_teller)   
            elif choice == 5: # 5. 입출금 업무
                view_process_balance(bank_teller)
            else:
                print('입력이 잘못 되었습니다. 메뉴 번호를 입력해주세요') 
                               
        except:
            print('입력이 잘못 되었습니다. 메뉴 번호를 입력해주세요')
 
 
if __name__ == '__main__':
    view_main()


#출처: https://uzooin.tistory.com/155 [뚱땡이 우주인]
```
