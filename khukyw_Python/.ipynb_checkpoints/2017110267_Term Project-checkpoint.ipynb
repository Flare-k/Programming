#### >>.dtypes로 앞으로 쓰게될 데이터들의 자료형을 알아보았다.

```python
#데이터 타입이 int로 돼 있는걸 str로 바꿔줌.
apt_price['연도'] = apt_price['연도'].astype(str)
apt_price['월'] = apt_price['월'].astype(str)
```

#### >>내용을 정리하기 쉽도록 '연도'와 '월'에 대해서 int에서 str타입으로 변환시켜주었다.

```python
apt_sale_price = apt_price['분양가격(㎡)']
```

```python
# '분양가격'이라는 없는 column을 숫자로 바꿔서 만들어서 넣어줌.
apt_price['분양가격'] = pd.to_numeric(apt_sale_price, errors = 'coerce')
# '평당분양가격'을 알고싶어서 계산하여 만들어줌.
apt_price['평당분양가격'] = apt_price['분양가격']*3.305785
```

### <연도에 따른 전국 규모별 평당분양가>  (단위: 천 원)

```python
#pivot_table 사용.
#pivot_table(value, index, column)
apt_price.pivot_table('평당분양가격', '규모구분','연도')
```

### <전국 평당분양가 연도별 변동금액>

```python
#'규모구분'이 '전체'인 평균으로 '평당분양가격'을 알아보았다.
#pivot_table(value, index, column)

total_year = apt_price.loc[apt_price['규모구분']=='전체']
total = total_year.pivot_table('평당분양가격', '지역명', '연도').reset_index()
total
```

```python
total['변동액'] = (total['2018'] - total['2015']).astype(int)
max_difference = np.max(total['변동액'])*1000  #변동이 가장 클 경우.
min_difference = np.min(total['변동액'])*1000  #변동이 가장 작을 경우.
mean_difference = np.mean(total['변동액'])*1000 #변동이 평균값일 경우.
total
#단위는 천 원.
```

### <30평 이상 규모에 대한 연도별 변동금액>  (단위: 천 원)

> **'전용면적 102㎡ 초과'(30평 이상)의 경우에 대해 변동금액을 알아보고 싶어서 위와 같은 내용으로 '규모 구분'만 '전용면적 102㎡ 초과'로 바꿔주었다.**

```python
total_year = apt_price.loc[apt_price['규모구분']=='전용면적 102㎡초과']
total = total_year.pivot_table('평당분양가격', '지역명', '연도').reset_index()
total
```

```python
total['변동액'] = (total['2017'] - total['2016']).astype(int)
max_difference = np.max(total['변동액'])*1000
min_difference = np.min(total['변동액'])*1000
mean_difference = np.mean(total['변동액'])*1000
total
#단위는 천 원.
```
