```python
input_var = ['id', 'hour', 'hour_bef_temperature', 'hour_bef_precipitation', 'hour_bef_windspeed', 'hour_bef_humidity', 'hour_bef_visibility', 'hour_bef_ozone', 'hour_bef_pm10', 'hour_bef_pm2.5']
target = 'count'
```

```python
rf.fit(train[input_var].fillna(0), train[target])
```

```python
import xgboost as xgb
trn_data = xgb.DMatrix(train[input_var], train[target])
```

```python
#파라미터값 설정
xgb_param = {'objective': 'reg:linear', #목적이 뭐냐
            'metric':'rmse', #어떤 metric을 쓸 것인지
            'max_depth':6, #각각의 DT의 최대 깊이
             'min_child_samples':10 #각각의 DT에 몇 개를 배정을 할 것인지.. 각 노드의 최소 샘플수 n개 밑으로 떨어지면 삭제됨
            }
```

```python
from sklearn.model_selection import KFold
folds = KFold(n_splits = 5, shuffle = True)
```

```python
test = pd.read_csv("test_final.csv")
```

```python
import numpy as np

final_predictions=np.zeros(len(test)) #test데이터에 대한 예측치를 넣어줄 공간을 만듬

for fold_, (trn_idx, val_idx) in enumerate(folds.split(train[input_var].values,train[target].values)):
    #enumerate는 해당인덱스를 추출하라는말.
    #folds로 train의 input_var과 target을 5번 스플릿
    trn_data = xgb.DMatrix(train[input_var].iloc[trn_idx],train[target].iloc[trn_idx])
    val_data = xgb.DMatrix(train[input_var].iloc[val_idx],train[target].iloc[val_idx])
    
    num_tree = 10000  #num_tree를 최대 10000까지 실행
    
    xgb_model = xgb.train(xgb_param,
                         trn_data,
                         num_tree,
                         [(trn_data,'train'),(val_data,'valid')], #trn_data를 통해 학습하고 val_data를 통해 예측 
                         early_stopping_rounds=200) #200이 넘어가는 순간까지 안 되면 멈춰라.
    final_predictions+=xgb_model.predict(xgb.DMatrix(test[input_var]), ntree_limit = xgb_model.best_ntree_limit) #가장 최적의 ntree를 구하라.

final_predictions/=5
```
