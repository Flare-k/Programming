```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd

df_keyword = pd.read_csv('keyword.csv')
keyword = df_keyword['Keyword']

driver = webdriver.Chrome("C:/Users/YHS/Google Drive/Webdriver/chromedriver")
driver.get("https://www.flipkart.com/")
driver.find_element_by_xpath('/html/body/div[2]/div/div/button').click()

words = []; results = []
for n in range(len(keyword)):
    print(keyword[n])
    driver.get("https://www.flipkart.com/")
    time.sleep(.5)
    driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input').send_keys(keyword[n])
    driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/button').click()
    time.sleep(1)

    source = driver.page_source
    soup = BeautifulSoup(source, 'html.parser')
    skew = soup.find_all('div', '_3wU53n')
    
    result = []; word = []
    for i in range(len(skew)):
        text = skew[i].get_text()
        word.append(keyword[n])
        result.append(text)
    words = words + word
    results = results + result
driver.close()
df = pd.DataFrame({'Keyword':words,
                    'Result':results})    
print("--- Crawled All ---")
```
