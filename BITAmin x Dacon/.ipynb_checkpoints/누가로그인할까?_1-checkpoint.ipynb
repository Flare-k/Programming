```python
train['Sex']=train['Sex'].map({'female':0, 'male':1})
test['Sex']=test['Sex'].map({'female':0, 'male':1})
```

```python
train['email_type'] = train['email_type'].map({'gmail':0, 
                                               'naver':1, 
                                               'other':2, 
                                               'nate':3, 
                                               'hanmail':4})
test['email_type'] = test['email_type'].map({'gmail':0, 
                                               'naver':1, 
                                               'other':2, 
                                               'nate':3, 
                                               'hanmail':4})
```

```python
train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=500, max_features = 6)  #n_estimators에 쓰이는 수는 해당 부트스트랩을 몇개나 만들것인지.. n개의 의사결정나무를 만들어서 평균을 낸다
#n_estimators -> 행의 다양성을 보장해줌..
#열의 다양성을 보장해주는 것도 필요하다..
#ex) (n_estimators = 500, max_features = 숫자)
# 6은 input_var의 70%정도.
```

```python
input_var = ['Sex', 'past_login_total', 'past_1_month_login','past_1_week_login','sub_size', 'email_type', 'phone_rat','apple_rat']
target = 'login'
```

```python
rf.fit(train[input_var].fillna(0), train[target])
```

```python
#xgboost 이용
#순차적인 부분을 다 설정해 줘야한다.

import xgboost as xgb
trn_data = xgb.DMatrix(train[input_var], train[target])
```

```python
#파라미터값 설정
xgb_param = {'objective': 'reg:linear', #목적이 뭐냐
            'metric':'rmse', #어떤 metric을 쓸 것인지
            'max_depth':6, #각각의 DT의 최대 깊이
             'min_child_samples':10 #각각의 DT에 몇 개를 배정을 할 것인지.. 각 노드의 최소 샘플수 n개 밑으로 떨어지면 삭제됨
            }
```

```python
from sklearn.model_selection import KFold
folds = KFold(n_splits = 5, shuffle = True)
```

```python
final_predictions=np.zeros(len(test)) #test데이터에 대한 예측치를 넣어줄 공간을 만듬

for fold_, (trn_idx, val_idx) in enumerate(folds.split(train[input_var].values,train[target].values)):
    #enumerate는 해당인덱스를 추출하라는말.
    #folds로 train의 input_var과 target을 5번 스플릿
    trn_data = xgb.DMatrix(train[input_var].iloc[trn_idx],train[target].iloc[trn_idx])
    val_data = xgb.DMatrix(train[input_var].iloc[val_idx],train[target].iloc[val_idx])
    
    num_tree = 10000  #num_tree를 최대 10000까지 실행
    
    xgb_model = xgb.train(xgb_param,
                         trn_data,
                         num_tree,
                         [(trn_data,'train'),(val_data,'valid')], #trn_data를 통해 학습하고 val_data를 통해 예측 
                         early_stopping_rounds=200) #200이 넘어가는 순간까지 안 되면 멈춰라.
    final_predictions+=xgb_model.predict(xgb.DMatrix(test[input_var]), ntree_limit = xgb_model.best_ntree_limit) #가장 최적의 ntree를 구하라.

final_predictions/=5
```

```python
xgb.train(xgb_param ,    #training을 하는 함수를 씀.
         trn_data,
          1000  #num_round는 이렇게 표현
         #num_round = 1000, #순차적으로 몇 번을 돌릴 것이냐
         #vervose_eval = #몇 번마다 출력이 되게 할 것인지
         )
#num_round의 숫자값의 차이는?? 1000과 10000의 차이는??
# >> 오차를 수정하는 행위를 1000번 할 것이냐 10000번 할 것이냐의 차이.
# 무한대로 할 이유가 없다. 왜?? 오버피팅(overfitting) 때문. 
# 즉, 지나치게 작은 에러를 가지고 진실이라고 판명하는 것 = Overfitting
# trn_data에 있는 쓸모없는 (운빨) 패턴까지 학습시킴 -> 성능 저하
```
